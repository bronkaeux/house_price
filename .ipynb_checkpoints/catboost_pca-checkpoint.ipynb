```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_df_2.csv')
```

```python
df.info()
```

```python
plt.figure(figsize=(20,16))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation heatmap for numerical features');
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
# Data to be predicted
df_to_predict = df[df['SalePrice'].isnull()]
df_to_predict = df_to_predict.drop(['SalePrice'], axis = 1)
```

```python
df = df.dropna()
```

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
len(df.columns)
```

```python
num_comp = range(1,200)
```

```python
explained_variance = []

for n in num_comp:
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(num_comp,explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained");
```

```python
pca_model = PCA(n_components=50)
```

```python
pca_X = pca_model.fit_transform(scaled_X)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_df = pd.DataFrame(data=pca_X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(pca_df, y, test_size=0.2)
```

```python
import catboost
```

```python
from catboost import CatBoostRegressor
```

```python
model = catboost.CatBoostRegressor(
    logging_level='Silent',
    iterations=2000,
    l2_leaf_reg=2,
    learning_rate= 0.03,
    depth = 5,
    random_seed = 42,
    eval_metric='RMSE',
    od_pval = 0,
    od_type='Iter', 
    od_wait=10
)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
r2_score(y_test, pred)
```

```python
MAE = mean_absolute_error(y_test, pred)
MSE = mean_squared_error(y_test, pred)
RMSE = np.sqrt(MSE)
```

```python
MAE
```

```python
RMSE
```

```python
test_residuals = y_test - pred
sns.scatterplot(x=y, y=test_residuals)
plt.axhline(y=0, color='red', ls='--');
```

```python
model.get_params()
```
